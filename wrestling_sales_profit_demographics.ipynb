```python
# Step 1: Import libraries
import pandas as pd
import random
from datetime import datetime, timedelta

# Step 2: Create sample data lists
promotions = ["NWA", "WWE", "AEW", "WCW", "AAA", "JCW", "ECW", "ROH"]
wrestlers = ["John Cena", "Kenny Omega", "Sting", "Rey Mysterio", "Sabu", "Chris Jericho", "Cody Rhodes", "Kazuchika Okada"]
countries = ["USA", "Canada", "Mexico", "Japan", "UK"]

match_types = ["Singles", "Tag Team", "Triple Threat", "Fatal 4-Way"]
events = ["Summer Slam", "WrestleMania", "All Out", "Bound for Glory", "Clash of Champions", "Revolution", "Final Battle"]

# Step 3: Generate sample dataset
data = []

for i in range(200):  # generate 200 rows
    promo = random.choice(promotions)
    w1 = random.choice(wrestlers)
    # pick a different opponent
    opp = random.choice([x for x in wrestlers if x != w1])
    event = random.choice(events)
    
    # random date
    date = datetime(2018, 1, 1) + timedelta(days=random.randint(0, 2000))
    
    # sales and profits
    tickets = random.randint(2000, 90000)
    ppv_sales = random.randint(50000, 2000000)
    profit = round(random.uniform(0.05, 0.5), 2)  # 5% to 50% profit
    
    # demographics
    age = random.randint(22, 55)
    country = random.choice(countries)
    
    data.append([
        promo, w1, opp, event, date.strftime("%Y-%m-%d"),
        random.choice(match_types), tickets, ppv_sales, profit,
        age, country
    ])

# Step 4: Create DataFrame
columns = [
    "Promotion", "Wrestler", "Opponent", "Event", "Date",
    "Match_Type", "Tickets_Sold", "PPV_Sales_USD", "Profit_Margin",
    "Wrestler_Age", "Country"
]

df = pd.DataFrame(data, columns=columns)

# Step 5: Inspect sample data
df.head()

# Step 6: Save to CSV
csv_filename = "wrestling_sales_demographics.csv"
df.to_csv(csv_filename, index=False)

print(f"CSV file saved as: {csv_filename}")
```
