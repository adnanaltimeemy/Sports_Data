```python
# Step 1 — Load libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import numpy as np
import pandas as pd

# Step 2 — Create a synthetic dataset
np.random.seed(42)

# Define store locations
stores = ['CityStore_Manchester', 'CityStore_NewYork', 'CityStore_AbuDhabi']

# Months (use 'ME' instead of deprecated 'M')
months = pd.date_range('2024-01-01', periods=12, freq='ME').month_name()

# Generate synthetic sales
data = []

for store in stores:
    for month in months:
        total_sales = np.random.randint(40000, 200000)  # total sales (£)
        profit_margin = np.random.uniform(0.1, 0.35)    # profit margin (10%–35%)
        profit = total_sales * profit_margin
        avg_age = np.random.randint(18, 60)             # average customer age
        region = np.random.choice(['Europe', 'North America', 'Middle East'])
        
        data.append([
            store, month, total_sales, profit, profit_margin, avg_age, region
        ])

# Create DataFrame
retail_df = pd.DataFrame(
    data,
    columns=[
        'Store',
        'Month',
        'Total_Sales',
        'Profit',
        'Profit_Margin',
        'Avg_Customer_Age',
        'Region'
    ]
)

# Display first few rows
retail_df.head()
```

```python
# Show head of dataset
retail_df.head()
```

```python
# Descriptive statistics
retail_df.describe()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='Store',
    y='Total_Sales',
    data=retail_df,
    estimator=np.mean,
    errorbar=None
)
plt.title('Average Monthly Sales by Store')
plt.ylabel('Avg Sales (£)')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Store', y='Profit_Margin', data=retail_df)
plt.title('Profit Margin Distribution by Store')
plt.ylabel('Profit Margin')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=retail_df, x='Avg_Customer_Age', hue='Region', multiple='stack')
plt.title('Customer Age Distribution by Region')
plt.xlabel('Avg Customer Age')
plt.show()
```
